```python
#remove link
def remove_url(text):
  pattern=re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

```python
def convert_chat_words(text):
  new_text=[]
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append((chat_words[w.upper()]).lower())
    else:
      new_text.append(w)
  return " ".join(new_text)
print(convert_chat_words(text))
```

```python
def remove_emoji(text):
  emoji_pattern=re.compile('['
                          u'\U0001F600-\U0001F64F'
                          u'\U0001F300-\U0001F5FF'
                          u'\U0001F680-\U0001F6FF'
                          u'\U0001F1E0-\U0001F1FF'
                          u'\U00002702-\U000027B0'
                          u'\U000024C2-\U0001F251'
                          ']+',flags=re.UNICODE)
  remaning_text=emoji_pattern.sub(r'',text)
  return remaning_text
text6='Hello are you happy😊'
print(remove_emoji(text6))
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stemming_words(text):
  new_text=[]
  for word in text.split():
    new_text.append(ps.stem(word))
  stem_text=' '.join(new_text)
  return stem_text
stemming_words('I am planning to do a work which is going to take a lot of time')
```
